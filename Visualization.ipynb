```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('rf_output.csv', nrows=200)
df.shape
```

```python
plt.figure(figsize=(12,5))
plt.xlabel('Samples')

s1 = pd.Series(df.pred)
s2 = pd.Series(df.actual)


ax1 = s1.plot(color='blue', grid=True, label='Prediction')
ax2 = s2.plot(color='red', grid=True, secondary_y=True, label='Actual')

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()


plt.legend(h1+h2, l1+l2, loc=2)
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(df.actual, df.pred)
```
